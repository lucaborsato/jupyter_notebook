# Read summary parameters of a fit and display properly here ... 

```python
import numpy as np
import pandas as pd
import os
```

```python
# function to load data and returns pandas frame
def load_data(file_in):
    skipl = 8
    cols = [0,1,2,5,6]
    names = [('par', '|S26'), ('unit','|S20'), 
             ('mle', 'f8'), ('nerr', 'f8'), ('perr', 'f8')
            ]
    d = np.genfromtxt(file_in, skip_header=skipl, usecols=cols, dtype=names)
    df = pd.DataFrame(d)
    return df
```

```python
file_par = os.path.abspath('/data2/borsato/TRADESsimulations/Kepler-9/EMCEE/2018-11-13_K9_emcee_noRV_001/2018-11-13_run_001/run_full/RV/0668_sim/parameters_summary.txt')
df = load_data(file_par)
```

```python
def display(df):
    
    nrow,ncol = np.shape(df)
    head = '%26s %20s %13s %13s %13s' %('parameter', 'unit', 'mle', '-1\sigma', '+1\sigma')
    print head
    for i in range(0,nrow):
        line = '%26s %20s %13.9f %13.9f %13.9f' %(df.par[i], df.unit[i],
                                                 df.mle[i], df.nerr[i], df.perr[i]
                                                 )
        print line
    return
```
